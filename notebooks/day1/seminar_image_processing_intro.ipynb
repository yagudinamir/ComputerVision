Загрузите изображение из файла tiger-border.png. Изображение состоит из рамки сплошного цвета и внутренней части изображения. Цвет рамки можно узнать, посмотрев на левый верхний пиксель. Рамка может иметь разную ширину со всех четырех сторон. Определите размеры рамки и выведите эти размеры через пробел. Размеры рамки выводите в следующем порядке: левый, верхний, правый, нижний.

```python
border = imread('tiger-border.png')
imshow(border)
```

```python
color = border[0, 0]
```

```python
left_border, up_border, right_border, down_border = [0, 0, 0, 0]
```

```python
xi, yi = 0, 1
x, y = int(border.shape[0] / 2), 0
while all(border[x, y] == color):
    x += xi
    y += yi
    left_border += 1
xi, yi = 1, 0
x, y = 0, int(border.shape[1] / 2)
while all(border[x, y] == color):
    x += xi
    y += yi
    up_border += 1

xi, yi = 0, -1
x, y = int(border.shape[0] / 2), border.shape[1] - 1
while all(border[x, y] == color):
    x += xi
    y += yi
    right_border += 1

    
xi, yi = -1, 0
x, y = border.shape[0] - 1, int(border.shape[1] / 2)
while all(border[x, y] == color):
    x += xi
    y += yi
    down_border += 1
    
print(left_border, up_border, right_border, down_border)
```
